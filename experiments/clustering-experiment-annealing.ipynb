# Clustering experiment for annealing

We will use the digits dataset for this experiment
(Num datapoints: `1797`, num features: `64`, num clusters: `10`)

```python
from mixes import GMN, DGMM, GMM, stopping_criterion
from sklearn import preprocessing, datasets, cluster
from sklearn.mixture import GaussianMixture
from utils import *
import seaborn as sns
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
data, labels = datasets.load_digits(return_X_y=True)
```

```python
alg_functions = [
    lambda e: GMN([10, 5, 2], [10, 6, 2], init='kmeans',
          update_rate=1, num_iter=100, evaluator=e,
          use_annealing=False,
          var_regularization=1e-3),
    lambda e: GMN([10, 5, 2], [10, 6, 2], init='kmeans',
          update_rate=1, num_iter=100, evaluator=e,
          use_annealing=True, annealing_start_v=0.5,
          var_regularization=1e-3),
    lambda e: GMN([10, 5, 2], [10, 6, 2], init='kmeans',
          update_rate=1, num_iter=100, evaluator=e,
          use_annealing=True, annealing_start_v=0.2,
          var_regularization=1e-3),
    lambda e: GMN([10, 5, 2], [10, 6, 2], init='kmeans',
          update_rate=1, num_iter=100, evaluator=e,
          use_annealing=True, annealing_start_v=0.1,
          var_regularization=1e-3)
]
alg_names = ["GMN", "GMN 0.5", "GMN 0.2", "GMN 0.1"]

result, evals = test_algorithms_on_data(alg_functions, alg_names,
                                   data, labels, num_repeats=20, return_evaluators=True)
```

```python
def plot_log_lik(evaluators, color=COLOR_BLUE, label=None):
    log_liks = [e.get_result_metric('log_lik') for e in evaluators]
    sns.kdeplot(x=log_liks, fill=False, color=color, label=label, bw_adjust=1)

fig = plt.figure(figsize=[6, 4], dpi=300)
plot_log_lik(evals["GMN"], color=COLOR_BLUE, label="1")
plot_log_lik(evals["GMN 0.5"], color=COLOR_GREEN, label="0.5")
plot_log_lik(evals["GMN 0.2"], color=COLOR_ORANGE, label="0.2")
plot_log_lik(evals["GMN 0.1"], color=COLOR_RED, label="0.1")
plt.legend()
plt.xticks(rotation = 0);
```
